```python


import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

standards = {
    'ph': {'ideal': 7.0, 'max': 8.5, 'min': 6.5},
    'Hardness': {'ideal': 150, 'max': 300},
    'Solids': {'ideal': 300, 'max': 600}, # Total Dissolved Solids
    'Chloramines': {'ideal': 2, 'max': 4},
    'Sulfate': {'ideal': 150, 'max': 250},
    'Conductivity': {'ideal': 500, 'max': 1000},
    'Organic_carbon': {'ideal': 5, 'max': 10},
    'Trihalomethanes': {'ideal': 50, 'max': 80},
    'Turbidity': {'ideal': 1, 'max': 5}
}

weights = {
    'ph': 0.10,
    'Hardness': 0.05,
    'Solids': 0.10,
    'Chloramines': 0.15,
    'Sulfate': 0.10,
    'Conductivity': 0.10,
    'Organic_carbon': 0.10,
    'Trihalomethanes': 0.15,
    'Turbidity': 0.15
}

# Ensure weights sum to 1
if not np.isclose(sum(weights.values()), 1.0):
    print("Warning: Weights do not sum to 1. Normalizing weights.")
    total_weight = sum(weights.values())
    weights = {key: value / total_weight for key, value in weights.items()}


def calculate_wqi(row, standards, weights):
    """
    Calculates a simplified Water Quality Index (WQI) for a single water sample.

    Args:
        row (pd.Series): A row from the water quality DataFrame.
        standards (dict): A dictionary containing ideal, min, and max standards for each parameter.
        weights (dict): A dictionary containing the weights for each parameter.

    Returns:
        float: The calculated WQI for the sample, or NaN if there's missing data
               or a parameter is not in the standards/weights.
    """
    wqi = 0
    for param, standard in standards.items():
        if param in row and param in weights:
            value = row[param]
            weight = weights[param]

            # Handle potential NaN values in the row
            if pd.isna(value):
                # print(f"Warning: Missing value for parameter '{param}'. Cannot calculate WQI.")
                return np.nan # Return NaN if any parameter is missing for WQI calculation

            # Calculate quality rating (qi) - This is a simplified example.
            # More sophisticated methods exist (e.g., based on curves, interpolation).
            # Here, we use a linear scaling approach.
            qi = 0
            if 'min' in standard and 'max' in standard:
                # For parameters with a range (like pH)
                ideal = standard['ideal']
                min_limit = standard['min']
                max_limit = standard['max']

                if value >= min_limit and value <= max_limit:
                    qi = 100 # Ideal quality within the standard range
                elif value < min_limit:
                    # Below minimum: quality decreases as value decreases
                    qi = max(0, 100 - (min_limit - value) / (min_limit - 0) * 100) # Linear decrease from 100 at min_limit to 0 at 0
                elif value > max_limit:
                    # Above maximum: quality decreases as value increases
                    qi = max(0, 100 - (value - max_limit) / (max_limit * 2 - max_limit) * 100) # Linear decrease from 100 at max_limit

            elif 'max' in standard:
                 # For parameters with only a maximum limit (like Hardness, Solids, etc.)
                ideal = standard.get('ideal', 0) # Assume ideal is 0 or a lower bound
                max_limit = standard['max']

                if value <= max_limit:
                    qi = 100 # Ideal quality up to the maximum limit
                else:
                    # Above maximum: quality decreases as value increases
                    # Linear decrease from 100 at max_limit to 0 at 2 * max_limit
                    qi = max(0, 100 - (value - max_limit) / (max_limit * 2 - max_limit) * 100)

            elif 'ideal' in standard:
                 # For parameters with only an ideal value (less common for WQI, but included for completeness)
                 # Quality decreases as value moves away from the ideal
                 ideal = standard['ideal']
                 # Simplified: Inverse relationship with deviation from ideal
                 deviation = abs(value - ideal)
                 qi = max(0, 100 - deviation / ideal * 100) # Linear decrease from 100 at ideal to 0 at 2*ideal (if ideal is not 0)

            else:
                print(f"Warning: Standard definition for '{param}' is incomplete. Cannot calculate quality rating.")
                return np.nan # Return NaN if standard is not defined correctly

            # Ensure qi is within 0-100 range
            qi = max(0, min(100, qi))

            # Weighted sum
            wqi += qi * weight
        else:
             # print(f"Warning: Parameter '{param}' not found in standards or weights.")
             pass # Silently ignore parameters not defined in standards/weights

    return wqi

# Load the dataset again to get the original dataframe with parameter names
df = pd.read_csv('/content/water_potability (1).csv')

# Impute missing values before calculating WQI
# Use the same imputer strategy as used for the models
imputer = SimpleImputer(strategy='mean')
df_imputed = pd.DataFrame(imputer.fit_transform(df.drop('Potability', axis=1)), columns=df.drop('Potability', axis=1).columns)
df_imputed['Potability'] = df['Potability'] # Add Potability back

# Calculate WQI for each row
df_imputed['WQI'] = df_imputed.apply(lambda row: calculate_wqi(row, standards, weights), axis=1)

# Display the dataframe with the calculated WQI
print(df_imputed[['ph', 'Hardness', 'Solids', 'Chloramines', 'Sulfate', 'Conductivity', 'Organic_carbon', 'Trihalomethanes', 'Turbidity', 'WQI', 'Potability']].head())

# Analyze the distribution of WQI
plt.figure(figsize=(10, 6))
sns.histplot(df_imputed['WQI'].dropna(), bins=50, kde=True)
plt.title('Distribution of Water Quality Index (WQI)')
plt.xlabel('WQI')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Analyze WQI vs Potability
plt.figure(figsize=(10, 6))
sns.boxplot(x='Potability', y='WQI', data=df_imputed)
plt.title('WQI Distribution by Potability')
plt.xlabel('Potability (0: Not Potable, 1: Potable)')
plt.ylabel('WQI')
plt.grid(True)
plt.show()


X_wqi = df_imputed[['WQI']].dropna() # Drop rows where WQI could not be calculated
y_wqi = df_imputed.loc[X_wqi.index, 'Potability'] # Get corresponding potability labels

X_train_wqi, X_test_wqi, y_train_wqi, y_test_wqi = train_test_split(X_wqi, y_wqi, test_size=0.2, random_state=42)


from sklearn.linear_model import LogisticRegression
model_wqi = LogisticRegression()
model_wqi.fit(X_train_wqi, y_train_wqi)
y_pred_wqi = model_wqi.predict(X_test_wqi)

accuracy_wqi = accuracy_score(y_test_wqi, y_pred_wqi)
print(f"\nAccuracy using WQI as a single feature: {accuracy_wqi}")
print(classification_report(y_test_wqi, y_pred_wqi))

cm_wqi = confusion_matrix(y_test_wqi, y_pred_wqi)
sns.heatmap(cm_wqi/np.sum(cm_wqi), annot=True, fmt='0.2%', cmap='Purples')
plt.title('Confusion Matrix using WQI')
plt.show()

# Note: This WQI calculation is a basic illustration. A proper WQI calculation requires
# defining appropriate standards, weights, and quality rating functions based on
# established water quality guidelines.
```

```python
print('Boxplot and density distribution of different features by Potability\n')

fig, ax = plt.subplots(ncols=2, nrows=9, figsize=(14, 28))

features = list(df.columns.drop('Potability'))

j = 0
for cols in features:
    # KDE Plot - Plotting each category separately as a workaround for ValueError with 'hue'
    temp_df = df[[cols, 'Potability']].dropna()

    sns.kdeplot(data=temp_df[temp_df['Potability'] == 0], x=cols, fill=True, alpha=0.4, color='indianred', label='0', ax=ax[j, 0])
    sns.kdeplot(data=temp_df[temp_df['Potability'] == 1], x=cols, fill=True, alpha=0.4, color='steelblue', label='1', ax=ax[j, 0])

    ax[j, 0].set_title(f'KDE Plot of {cols} by Potability')
    ax[j, 0].set_xlabel(cols)
    ax[j, 0].set_ylabel('Density')
    ax[j, 0].legend(title='Potability') # Add legend for the two categories

    # Boxplot
    sns.boxplot(x=df['Potability'], y=df[cols], palette=['indianred', 'steelblue'], ax=ax[j, 1])
    ax[j, 1].set_title(f'Boxplot of {cols} by Potability')
    ax[j, 1].set_xlabel('Potability')
    ax[j, 1].set_ylabel(cols)

    j += 1

plt.tight_layout()
plt.show()
```
